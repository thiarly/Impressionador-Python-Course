## 1 - Importação da base de dados e tratamento

```python
import pandas as pd
funcionarios_df = pd.read_csv('CadastroFuncionarios.csv', sep=';', decimal=',')
clientes_df = pd.read_csv('CadastroClientes.csv', sep=';', decimal=',')
servicos_df = pd.read_excel('BaseServiçosPrestados.xlsx', decimal=',')

# remoção estado civil e cargo

funcionarios_df = funcionarios_df.drop(['Estado Civil', 'Cargo'], axis=1)

display(funcionarios_df)
display(clientes_df)
display(servicos_df)

```

## 2 - Valor Total da Folha Salarial 

```python
#Criando a columa salario total e fazendo a soma 
funcionarios_df['Salario Total'] = funcionarios_df['Salario Base'] + funcionarios_df['Impostos'] + funcionarios_df['Beneficios'] + funcionarios_df['VT'] + funcionarios_df['VR']
print('Total da folha salário R$ {:,}'.format(funcionarios_df['Salario Total'].sum()))
```

## 3 - Qual foi o faturamento da empresa?

```python
faturamento_df = servicos_df[['ID Cliente', 'Tempo Total de Contrato (Meses)']].merge(clientes_df[['ID Cliente', 'Valor Contrato Mensal' ]])
faturamento_df['Faturamento Total'] = faturamento_df['Tempo Total de Contrato (Meses)'] * faturamento_df['Valor Contrato Mensal']
print('Faturamento total foi de R$ {:,}'.format(sum(faturamento_df['Faturamento Total'])))
```

## 4 - Qual o % de funcionários que já fechou algum contrato?

```python
#.unique() remove os valores de ID duplicados

qtde_funcionario_fechamentocontrato = len(servicos_df['ID Funcionário'].unique())
qtde_funcionario_total = len(funcionarios_df['ID Funcionário'])

print('Percentual de funcionários que fecharam contrato {:.2%}'.format(qtde_funcionario_fechamentocontrato / qtde_funcionario_total))
```
