### Vamos analisar o IBOV

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader.data as web
import yfinance as yf

cotacao_ibov = yf.download('^BVSP', start="2010-01-01", end=None)
display (cotacao_ibov)

# cotacao_ibov = web.DataReader('^BVSP', data_source='yahoo', start='2020-1-1', end='2020-11-30')
# display(cotacao_ibov)

```

### Analisando o Gráfico

```python
cotacao_ibov['Adj Close'].plot(figsize=(15, 5))
```

### Retorno do IBOV

```python
retorno_ibov = cotacao_ibov['Adj Close'][-1] / cotacao_ibov['Adj Close'][0]-1
print('{:.2%}'.format(retorno_ibov))
```

### Analisando com Média Móvel

```python
cotacao_ibov['Adj Close'].plot(figsize=(15, 5), label='IBOV')
cotacao_ibov['Adj Close'].rolling(21).mean().plot(label='M21')
cotacao_ibov['Adj Close'].rolling(34).mean().plot(label='M34')
plt.legend()
plt.show()
```
